```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("supply_chain_data.csv")
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['Profit'] = df['Revenue generated'] - df['Costs']
```

```python
df.groupby('Transportation modes')['Revenue generated'].sum()
```

```python
df.groupby('Transportation modes')['Costs'].mean()
```

```python
df['Shrinkage'] = df['Stock levels'] - df['Number of products sold']
print(df['Shrinkage'])
```

```python
df.groupby('Manufacturing lead time')['Production volumes'].mean()
```

```python
df.groupby('Manufacturing lead time')['Production volumes'].mean()
```

```python
df.groupby('Production volumes')['Manufacturing costs'].mean()
```

```python
df.groupby('Supplier name')['Defect rates'].mean()
```

```python
df.groupby('Supplier name').agg({
    'Manufacturing costs': 'mean',
    'Defect rates': 'mean',
    'Lead time': 'mean'
})
```

```python
df['Inspection results'].value_counts()
```

```python
supplier_perf = df.groupby('Supplier name').agg({
    'Manufacturing costs': 'mean',
    'Defect rates': 'mean',
    'Lead time': 'mean'
}).reset_index()

supplier_perf['Score'] = (
    supplier_perf['Manufacturing costs'] +
    supplier_perf['Defect rates'] +
    supplier_perf['Lead time']
)

top5 = supplier_perf.nsmallest(5, 'Score')
worst5 = supplier_perf.nlargest(5, 'Score')
```
